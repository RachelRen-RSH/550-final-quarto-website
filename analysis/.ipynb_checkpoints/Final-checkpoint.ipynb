```python
page_results = []
results = []
for div in divs:
    name = div.find('h3', class_='uitk-heading uitk-heading-5 overflow-wrap uitk-layout-grid-item uitk-layout-grid-item-has-row-start').text

    general_location = div.find('div', class_='uitk-text uitk-text-spacing-half truncate-lines-2 uitk-type-300 uitk-text-default-theme').text

    amenities = div.find_all('div', class_ = 'uitk-text truncate-lines-2 uitk-type-200 uitk-text-default-theme') 
    amenity_text = ""
    for amenity in amenities:
        amenity_text += amenity.text+"; "
    amenity_text = amenity_text[:-2] 
       
    ## For ratings, there are two conditions: one high rating and one medium rating, which have different class names. In order to 
    ## avoid the "None" situation, we create a loop and if condition statement.
    classes_to_search = [
    "uitk-badge uitk-badge-base-large uitk-badge-base-has-text uitk-badge-positive",
    "uitk-badge uitk-badge-base-large uitk-badge-base-has-text uitk-badge-standard"
    ]

    # Find the parent span with the specified classes
    rating = '0' 
    for class_name in classes_to_search:
        parent_span = div.find('span', class_=class_name)
        if parent_span:
            # If found, extract the rating
            rating_span = parent_span.find('span', class_='uitk-badge-base-text')
            if rating_span:
                rating = float(rating_span.text.strip())
                break
    
    price = div.find('div', class_='uitk-text uitk-type-end uitk-type-200 uitk-text-default-theme').text
    price = int(price.replace("$", "").replace(" total", "").replace(",", "").strip())
    # Save the result
    page_results.append([name, general_location, amenity_text, rating, price])
page_results

# Create a dataframe and save
col_names = ["name", "general_location", "amenities", "rating", "price"]
df = pd.DataFrame(page_results, columns=col_names)
results.append(df)
results = pd.concat(results, axis=0).reset_index(drop=True)
results.index.name = 'ID'
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
def get_xy_from_geometry(df):
    x = df.geometry.centroid.x
    y = df.geometry.centroid.y
    return np.column_stack((x, y))
```

```python
hotelsXY = get_xy_from_geometry(hotels_census_joined_3857)
park_url = "https://gisweb.miamidade.gov/arcgis/rest/services/Parks/MD_Parks305/MapServer/0/query?outFields=*&where=1%3D1&f=geojson"
parks = gpd.read_file(park_url)
# Get the X/Y
parksXY = get_xy_from_geometry(parks.to_crs(epsg=3857))
# # Run the k nearest algorithm
nbrs = NearestNeighbors(n_neighbors=1)
nbrs.fit(parksXY)
parksDists, _ = nbrs.kneighbors(hotelsXY)

# # Add the new feature
hotels_census_joined_3857["logDistParks"] = np.log10(parksDists.mean(axis=1))
```

```python
transit_url = "https://services.arcgis.com/8Pc9XBTAsYuxx9Ny/arcgis/rest/services/MetroRailStations_gdb/FeatureServer/0/query?outFields=*&where=1%3D1&f=geojson"
transit_stops = gpd.read_file(transit_url)
# Get the X/Y
transitXY = get_xy_from_geometry(transit_stops.to_crs(epsg=3857))
# # Run the k nearest algorithm
transit_nbrs = NearestNeighbors(n_neighbors=1)
transit_nbrs.fit(transitXY)
transitDists, _ = transit_nbrs.kneighbors(hotelsXY)

# # Add the new feature
hotels_census_joined_3857["logDistMetro"] = np.log10(transitDists.mean(axis=1))
```

```python
Miami_311 = pd.read_csv('311_Service_Requests_-_Miami-Dade_County_-_2023.csv')
trash_complaint = Miami_311[Miami_311['issue_type'] == 'BULKY TRASH COMPLAINT']
trash_complaint.head()
trash_complaint_gdf = gpd.GeoDataFrame(
    trash_complaint, 
    geometry=gpd.points_from_xy(trash_complaint['longitude'], trash_complaint['latitude'])
)
trash_complaint_gdf = trash_complaint_gdf.set_crs(epsg=4326)

trashXY = get_xy_from_geometry(trash_complaint_gdf.to_crs(epsg=3857))
trashXY = trashXY[~np.isnan(trashXY).any(axis=1)]
len(trashXY)

trash_nbrs = NearestNeighbors(n_neighbors=5)
trash_nbrs.fit(trashXY)
trashDists, _ = trash_nbrs.kneighbors(hotelsXY)

# Handle any sales that have 0 distances
trashDists[trashDists == 0] = 1e-5  # a small, arbitrary value

# Add the new feature
hotels_census_joined_3857["logDistTrash"] = np.log10(trashDists.mean(axis=1))
```

```python
abandoned = Miami_311[Miami_311['issue_type'] == 'ABANDONED PROPERTY / VEHICLE']

abandoned_gdf = gpd.GeoDataFrame(
    abandoned, 
    geometry=gpd.points_from_xy(abandoned['longitude'], abandoned['latitude'])
)
abandoned_gdf = abandoned_gdf.set_crs(epsg=4326)

abandonedXY = get_xy_from_geometry(abandoned_gdf.to_crs(epsg=3857))
abandonedXY = abandonedXY[~np.isnan(abandonedXY).any(axis=1)]
len(abandonedXY)

abandoned_nbrs = NearestNeighbors(n_neighbors=5)
abandoned_nbrs.fit(abandonedXY)
abandonedDists, _ = abandoned_nbrs.kneighbors(hotelsXY)

# Handle any hotels that have 0 distances
abandonedDists[abandonedDists == 0] = 1e-5  # a small, arbitrary value

# Add the new feature
hotels_census_joined_3857["logDistAbandoned"] = np.log10(abandonedDists.mean(axis=1))
```
